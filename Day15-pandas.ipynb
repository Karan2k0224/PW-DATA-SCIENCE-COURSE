A DataFrame is a two-dimensional, tabular data structure provided by the Pandas library.
It’s similar to a spreadsheet or a SQL table, where data is organized into rows and columns.
Each column in a DataFrame can have a different data type (e.g., numeric, string, datetime).
You can think of a DataFrame as a collection of Series objects, where each Series represents a column.

Q5. What are some common functions you can use to manipulate data in a Pandas DataFrame? Can
you give an example of when you might use one of these functions?

1)head() and tail():
These functions allow you to view the first (head) or last (tail) few rows of your DataFrame.
Useful for quickly inspecting dat

```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")

df.head()  # Show the first 5 rows
df.tail(3)  # Show the last 3 rows
#type(df)
```

2->describe():
 Generates summary statistics for numeric columns (mean, min, max, etc.).
 Helps understand the distribution of data.



```python
df.describe()
```

3-->info():
Provides information about the DataFrame, including data types and non-null counts.
Useful for checking missing values.

```python
df.info()
```

4-->groupby():
Groups data based on one or more columns.
Allows aggregation (sum, mean, count, etc.) within each group.

```python
df.groupby('Ticket').count()
# df.groupby('Ticket')["Pclass"].sum()
```

5-->sort_values():
Sorts the DataFrame based on specified columns.
Useful for arranging data in a desired order.

```python
# df.sort_values(by='Date', ascending=True)
```

6-->apply():
Applies a function to each element or row/column of the DataFrame.
Useful for custom transformations.

```python
def double_value(x):
    return 2 * x

df['Doubled_Sales'] = df['Fare'].apply(double_value)
df
```

7-->fillna():
Fills missing values (NaN) with specified values (mean, median, etc.).
Useful for data imputation

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
df
```

Q7-->Create a DataFrame using multiple Series. Explain with an example.

 Pandas DataFrame using multiple Series. A DataFrame is a two-dimensional tabular data structure, and each column in a DataFrame can be represented as a Series. 

```python
import pandas as pd

# Create individual Series for each subject
math_scores = pd.Series([85, 90, 78, 92], name="Math Scores", index=["Alice", "Bob", "Carol", "David"])
english_scores = pd.Series([88, 82, 95, 89], name="English Scores", index=["Alice", "Bob", "Carol", "David"])
history_scores = pd.Series([75, 80, 88, 92], name="History Scores", index=["Alice", "Bob", "Carol", "David"])

# Combine the Series into a DataFrame
student_performance_df = pd.DataFrame({
    "Math": math_scores,
    "English": english_scores,
    "History": history_scores
})

print(student_performance_df)
```
